The Lorenz equations are a set of 3 autonomous ordinary differential equations.

$$ \begin{align}
    \dot{x} &= \sigma (y - x) \\
    \dot{y} &= x (r - z) - y \\
    \dot{z} &= x y - bz
\end{align} $$


Or in matrix form:

$$ \begin{pmatrix}
\dot{x} \\
\dot{y} \\
\dot{z}
\end{pmatrix} = 
\begin{pmatrix}
- \sigma & \sigma & 0 \\
r - z & -1 & - x \\
y & x & -b
\end{pmatrix}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}$$

```python
def lorenz(xyz, sigma, r, b):
    x, y, z = xyz
    xdot = sigma * (y - x)
    ydot = x * (r - z) - y
    zdot = x * y - b * z
    return np.array([xdot, ydot, zdot])

def fix_lorenz_params(sigma, r, b):
    """Fix the lorenz attractor with set of parameters.
       This can then be passed to a numerical integration method."""
    @wraps(lorenz)
    def fixed_lorenz(x):
        return lorenz(x, sigma, r, b)
    return fixed_lorenz


sigma = 10.0
r =     28.0
b =     8.0/3.0
STD_LORENZ = fix_lorenz_params(sigma, r, b)
```

### Numerical Integration methods

#### Double Approximation

This is the method that Lorenz used in his original study of the equations.

### RK4

The Runge-Kutta 4th order method.



```python
def forward_euler(x,h,f):
    """Forward Euler integration method for an autonomous system f."""
    return x + h * f(x)

def RK4(x, h, f):
    """Runge-Kutta 4th-order integration method for an autonomous system f."""
    K1 = h*f(x)
    K2 = h*f(x+(K1 * 0.5))
    K3 = h*f(x+(K2 * 0.5))
    K4 = h*f(x+K3)
    return x+(1/6.0)*(K1 + 2*K2 + 2*K3 + K4)

def double_approx(x, h, f):
    """The method used by E. Lorenz in his original study of the equations."""
    x1 = forward_euler(x, h, f)
    x2 = forward_euler(x1, h, f)
    return 0.5*(x + x2)
```

```python
def plot_points(points, dimensions=('x y', 'x z', 'y z'), chartsize=(5,4)):
    """Plot a set of 3D points as a list of graphs using the given dimensions as XY coordinates."""
    points = np.array(points)
    plt.figure(figsize=(len(dimensions)*chartsize[0], chartsize[1]))
    for i, axes in (enumerate(x.split() for x in dimensions)):
        if axes[0] is 'n':
            x = range(len(points))
            y = points[:,{'x':0, 'y':1, 'z':2}[axes[1]]]
        else:
            x, y = [points[:,{'x':0, 'y':1, 'z':2}[a]] for a in axes]
        plt.subplot('1%d%d'% (len(dimensions), i+1))
        plt.xlabel(axes[0])
        plt.ylabel(axes[1])
        plt.plot(x, y)
        
def run_lorenz(x0, sigma, r, b, h=0.01, n=6000, maxt=None, integrator=RK4, return_ts=False):
    """Performs numerical integration of the Lorenz equations for the specified parameters."""
    x = x0
    l = fix_lorenz_params(sigma, r, b)
    if maxt:
        n = int(maxt / h) - 1
    pos = [x]
    for i in tqdm(range(n)):
        x = integrator(x, h, l)
        pos.append(x)
    if return_ts:  # return the time points as well as postions
        return np.linspace(0, t, maxt/h), np.array(pos)
    return np.array(pos)
```

### Lyapunov Exponents

```python
# Gram-Schmidt Orthonormalisation

def normalise(u):
    return u / vec_abs(u)

def project(u, v):
    return u * (np.dot(u, v) / np.dot(u, u))

def gram_schmidt(vs, normalised=True):
    """Gram-Schmidt Orthonormalisation / Orthogonisation.
    Given a set of vectors, returns an orthogonal set of vectors
    spanning the same subspace.
    Set `normalised` to False to return a non-normalised set of vectors."""
    us = []
    for v in np.array(vs):
        u = v - np.sum([project(x, v) for x in us], axis=0)
        us.append(u)
    if normalised:
        return np.array([normalise(u) for u in us])
    else:
        return np.array(us)

    
def gram_schmidt_modified(vs, normalised=True):
    """Gram-Schmidt Orthonormalisation / Orthogonisation.
    Given a set of vectors, returns an orthogonal set of vectors
    spanning the same subspace.
    Set `normalised` to False to return a non-normalised set of vectors.
    
    (Modified version as per p.80 Practical Numerical Algorithms 
        - Parker & Chua)"""
    us = []
    for v in np.array(vs):
        u = v
        for x in us:
            u = u - project(x, u)
        us.append(u)
    if normalised:
        return np.array([normalise(u) for u in us])
    else:
        return np.array(us)

    
## testing  Gram-Schmidt works as expected
print("Testing Gram-Schmidt orthonormalisation.")

randmat = np.random.random((3,3))
print("Random matrix:")
print(randmat)
gsrm = gram_schmidt(randmat)
print("Normalised:")
print(gsrm)
print("x1 . x2 = ", np.dot(gsrm[0], gsrm[1]))
print("x1 . x3 = ", np.dot(gsrm[0], gsrm[2]))
print("x2 . x3 = ", np.dot(gsrm[1], gsrm[2]))
print("x2 . x2 = ", np.dot(gsrm[1], gsrm[1]))
```

```python
def augment_position(x, u):
    """Combine phase-space position and linearised vector matrix."""
    return np.append([x], u, axis=0)

def unaugment_position(xu):
    """Break apart an augmented position matrix into phase-space position and perturbation matrix."""
    return (xu[0], xu[1:])

def extended_lorenz(xu, sigma=10.0, r=28.0, b=8.0/3.0):
    """Extended Lorenz system includes the linearised equations.
    Given a 4 vector list `xu`, where row 1 is the position in
    phase space and rows 2-4 are 3 perturbation vectors.
    Returns 4 vector list of the same format for the new position."""
    xyz, u = unaugment_position(xu)
    x,y,z = xyz
    xyz_ = lorenz(xyz, sigma, r, b)
    u_ = []
    for dx, dy, dz in u:
        dxdot = -sigma*dx + sigma*dy
        dydot = (r - z) * dx - dy - x * dz
        dzdot = y * dx + x * dy - b * dz
        u_.append([dxdot, dydot, dzdot])
    return augment_position(xyz_, u_)

def lyapunov_exponents(x0, u0, T, h, f):
    """Numerically calculate the Lyapunov Exponents of the augmented
    nonlinear system f.
    x0:  Initial Position in phase space.
    u0:  A list of initial perturbation vectors.
    f:  The system of nonlinear and linearised equations.
    Returns a list of Lyapunov Exponents."""
    x, u = x0, u0
    sums= [0,0,0]
    for i in tqdm(range(int(T/h))):
        xu = augment_position(x, u)
        xu1 = RK4(xu, h, extended_lorenz)
        x, u = unaugment_position(xu1)
        v = gram_schmidt(u, normalised=False)
        u = gram_schmidt(v)
        sums += np.log(vec_abs(v))
    return sums / T

def kaplan_yorke_dimension(lyapunov_exponents):
    """Calculates the Kaplan Yorke Dimension of a list of Lyapunov exponents.
        D = k + sum[i = 1 to k](exponent[i] / abs(exponent[k+1])
    where k is the largest number of exponents that can be summed together to a total > 0.
    """
    lexp = np.sort(lyapunov_exponents)[::-1]
    k = np.sum(lexp.cumsum() > 0)
    return k + np.sum([e / np.abs(lexp[k]) for e in lexp[:k]])
        
```

```python
T =  10000
h =  0.001
x0 = x = np.array([5.2, 8.5, 27.0])
u0 = np.identity(3)

lyxp = lyapunov_exponents(x0, u0, T, h, extended_lorenz)

print("Lyapunov Exponents:")
for i, l in enumerate(lyxp):
    print("λ%d: %.4f"% (i+1, l))

# Lyapunov Exponents should sum to divergence  # (-13.6666...)
print("Sum of exponents: %.5f (should match divergence: ∇ · u = -13.666...)" % sum(lyxp))

print("Kaplan-Yorke dimension: %.3f" % kaplan_yorke_dimension(lyxp))
```

### Boundedness of the Lorenz system

Consider the divergence of the system:

$$ \nabla \cdot \vec{L} = d\dot{x} $$

### Fixed Points of the system

$$
\begin{pmatrix}
- \sigma & \sigma & 0 \\
r - z & -1 & - x \\
y & x & -b
\end{pmatrix}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
=
\begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}$$

The equations have a fixed point at $ (0,0,0) $ at all values of $ \sigma, r, b $.  There are also two more fixed points:

$$
\begin{align}
  \sigma ( y - x ) = 0 &\implies x = y \\
  x y - b z = 0 &\implies x^{2} - b z = 0 \implies z = \frac{x^{2}}{b} \\
  x ( r - z) - y = 0 &\implies x ( r - \frac{x^{2}}{b} ) - x = 0 \\
  &\implies r - \frac{x^{2}}{b} - 1 = 0 \quad \text{for }  x \neq 0 \\
  &\implies x = \pm\sqrt{b ( r - 1 )}
\end{align}
$$

Thus there are two additional fixed points at  $ ( \pm\sqrt{b ( r - 1 )}, \pm\sqrt{b ( r - 1 )}, r - 1 ) $.  These are dependent on parameter $ r $.  $ r $ is a the *Rayleigh Number*, a positive real number.


### Analytic Bifurcation of r

Given the fixed point equation above, we can examine the bifurcation of x and y position as function of r:

```python
def r_bifurcation(b, r):
    if r <= 1:
        return 0.0  # we're only interested in real results, for r < 1 the result is imaginary
    return np.sqrt(b * (r - 1))

b = 8.0/3.0
r_range = np.linspace(0, 35, 200)
plt.plot(r_range, [r_bifurcation(b, r) for r in r_range], color='grey')
plt.plot(r_range, [-r_bifurcation(b, r) for r in r_range], color='grey')
plt.plot(r_range, 0*r_range, color='grey', linestyle='--')
plt.xlabel('r')
plt.ylabel('x')
save_chart('bifurcation')
```

```python
### Poincare Return Map
def return_map_crossings(trajectory, z=None):
    """Returns a list of pairs of points where the trajectory crosses the z plane."""
    if z is None:
        # take the z-position of the first point of the trajectory
        z = trajectory[0][2]
    # find sets of consecutive points where the z value crosses the given plane `z`
    consecutive_xs = zip(trajectory[:-1], trajectory[1:]) 
    intersections = [(x1, x2) for x1, x2 in consecutive_xs if x1[2] >= z and x2[2] < z]
    return np.array(intersections)

def return_map(trajectory, z=None):
    """Returns a list of points where the trajectory crosses the z plane.
    Uses linear interpolation to find precise intersections with the
    transverse plane."""
    if z is None:
        z = trajectory[0][2]
    intersections = return_map_crossings(trajectory, z)
    # linear interpolation between the two points to get a value on the given plane `z`
    returns = [x1+(z - x1[2]) / (x2[2]-x1[2])*(x2-x1) for x1, x2 in intersections]
    return np.array(returns)


# show the return map
returns_interp = return_map(points)
plt.scatter(returns_interp[:,0], returns_interp[:,1])
plt.scatter(-returns_interp[:,0], -returns_interp[:,1], color='red')
plt.scatter([fixed_point, -fixed_point], [fixed_point, -fixed_point], color='green')
```

#### Mark the returns by the side of the attractor they went around

critical points $C_\pm$ lie at $ ( \pm\sqrt{b ( r - 1 )}, \pm\sqrt{b ( r - 1 )}, r - 1 ) $ so we expect each return to go around these points before returning to $\Sigma$.

For the classical values of Lorenz, that places the fixed points at $\left(\pm 8.4853, \pm 8.4853, 27 \right)$, so we can mark each orbit by seeing whether it goes to an value more or less than $\pm 8.4853$.

```python
# Plot the return map and mark each point by the fixed point
# that it orbits after leaving the return plane

MINUS, PLUS = -1,1

# determine which fixed point the trajectory last orbited
wing = MINUS
wings = []
for x,y,z in points:
    if x < -fixed_point and y < -fixed_point and z > (r - 1):
        wing = MINUS
    if x > fixed_point and y > fixed_point and z > (r - 1):
        wing = PLUS
    wings.append(wing)

# match the return map to the last orbit
tagged_returns = list(zip(returns_interp, [w for p, w in zip(points[1:], wings) if p in sigma_intersection[:,0]]))
minus_wing = np.array([p for (p,w) in tagged_returns if w is MINUS])
plus_wing = np.array([p for (p,w) in tagged_returns if w is PLUS])

# plot on a square plane
plt.figure(figsize=(8,8))
ax = plt.subplot(111)
plt.scatter(plus_wing[:,0], plus_wing[:,1], color='blue', label='$C_-$ orbit: leaves $\Sigma$ along $\gamma_-$')
plt.scatter(minus_wing[:,0], minus_wing[:,1], color='red', label='$C_+$ orbit: leaves $\Sigma$ along $\gamma_+$')
plt.scatter(-plus_wing[:,0], -plus_wing[:,1], color='red')
plt.scatter(-minus_wing[:,0], -minus_wing[:,1], color='blue')
plt.scatter([fixed_point, -fixed_point], [fixed_point, -fixed_point], color='green', marker='+')
plt.text(fixed_point+.2, fixed_point+.2, '$C_+$')
plt.text(-fixed_point-.5, -fixed_point-.5, '$C_-$')
plt.text(-points[0,0]+.2, -points[0,1]+.2, '$x_0$')
plt.text(points[0,0]-.5, points[0,1]-.5, '$x_0$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')

# draw the stable manifold between the two branches
manifold = np.array([
    (min(plus_wing, key=lambda x: x[1]) + max(minus_wing, key=lambda x: x[0]))/2,
    (max(minus_wing, key=lambda x: x[1]) + min(plus_wing, key=lambda x: x[0]))/2
    ])

m = (manifold[1,1] - manifold[0,1]) / (manifold[1,0] - manifold[0,0])
c = manifold[0,1] - manifold[0,0]*m
edges = [x*m+c for x in [-6,6]]
plt.plot([-6,6], edges, color='black', linestyle='--')
plt.text(6.4,-3, '$\Gamma$')

save_chart('first_return_map')
```

```python
## Time-step halving  p. 50 Practical Numerical Algorithms (Parker & Chua)

sigma = 10.0
r = 28.0
b = 8.0/3.0
h=0.001
high_prec_points = run_lorenz(np.array([-5.2, -8.5, r -1]), sigma, r, b, h=h, n=600000)
plot_points(high_prec_points)
```

```python
def time_half_crossings(crossings, z, epsilon=1e-6, h=0.001):
    """Use the time-step halving algorithm to intergrate forward with high precision onto the
    return map z plane [p. 50 Practical Numerical Algorithms (Parker & Chua)].
    Integrates forward from x1 until intersecting the return plane.  If
    the plane is not hit within a given accuracy, half the time-step size and repeat from x1.
    Given a list of pairs of points that cross the return plane,
    returns a list of trajectory intersections with the z plane with a
    tolerance of |z position - map z plane| < epsilon."""
    intersections = []
    for x1, x2 in tqdm(crossings):
        # x* where z = r - 1 must lie between x1 and x2, so continue to halve the time step and integrate forward
        # until a desired tolerance is achieved
        h_ = h
        accurate_enough = False
        while not accurate_enough:
            h_ = h_ / 2.0
            x_ = x1
            while x_[2] > z:
                x_1 = x_
                x_ = RK4(x_, h_, STD_LORENZ)
            if np.abs(x_[2] - z) < epsilon:
                intersections.append(x_)
                accurate_enough = True
            elif np.abs(x_1[2] - z) < epsilon:
                intersections.append(x_1)
                accurate_enough = True
    return np.array(intersections)
```

```python
## WARNING:  This may take a long time to run (high_prec_points has a lot of intersections)
crossings = return_map_crossings(high_prec_points)[1:]  # ignore x0, we know it's on r - 1
intersections = time_half_crossings(crossings, (r-1), epsilon=0.000001)

plt.figure(figsize=(12,12))
ax = plt.subplot(111)
plt.scatter(intersections[:,0], intersections[:,1], color='grey', marker='.')
plt.scatter(-intersections[:,0], -intersections[:,1], color='grey', marker='.')
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim(-4, -3)
ax.set_ylim(-.5, .5)
```

### Map of the interval

To show sensitivity to initial conditions, the map of the interval x -> f(x) should be expanding (f' > 1)

```python
# expanding map of the interval

def first_return(x0, z=None, lorenz=STD_LORENZ):
    """Calculate the first return point to the given z-plane from an initial point."""
    if z is None:
        z = x0[2]
    x_1, x = x0, RK4(x0, 0.01, lorenz)
    xs = [x_1, x]
    # assume that trajectory always leaves the z-plane going down first
    # (true for Lorenz equations near the stable manifold, xy < bz)
    while not (x_1[2] > z and x[2] <= z):
        x_1, x = x, RK4(x, 0.01, lorenz)
        xs.append(x)
    return return_map(xs, z)[1]

interval_map = []
for i in tqdm(np.append(np.arange(-10, 0, 0.01), np.arange(0.01, 10.01, 0.01))):
    # single loop for a given x0 position
    x0 = np.array([i, 0.0, 27.0])
    interval_map.append(np.array([x0, first_return(x0)]))

interval_map = np.array(interval_map)
plt.plot(interval_map[:1000,0,0], interval_map[:1000,1,0])
plt.plot(interval_map[1000:,0,0], interval_map[1000:,1,0])
```

## Warwick Tucker's change of coordinates

W. Tucker made a change of coordinates using a Jordan normal form of the system to align the stable and unstable manifolds of the system with the coordinate system.

$$\begin{align}
\dot{x} &= \lambda_1 x - k_1(x +y)z \\
\dot{y} &= \lambda_2 y + k_1(x +y)z \\
\dot{z} &= \lambda_3 z + k_1(x +y)(k_2 x + k_3 y)
\end{align}$$

$$\begin{align}
k_1 &= 0.29 \\
k_2 &= 2.2 \\
k_3 &= 1.3
\end{align}$$

```python

def lorenz_j(xyz):
    x, y, z = xyz
    xdot = 11.8*x - 0.29 * (x + y) * z
    ydot = -22.8*y + 0.29 * (x + y) * z
    zdot = -2.67*z + (x + y) * (2.2*x - 1.3*y)
    return np.array([xdot, ydot, zdot])

x0 = np.array([0.0, 1.0, 0.0])
x = x0
xs = [x]
for i in tqdm(range(60000)):
    x = RK4(x, 0.001, lorenz_j)
    xs.append(x)
    
plot_points(xs)
```
