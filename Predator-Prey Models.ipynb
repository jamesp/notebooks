# Predator-Prey models

## Rabbits

A population of rabbits (R) grows by a rate r:

$$ r = birthrate - deathrate $$

$$ \frac{dR}{dt} = rR $$

While can easily solve this to get the rabbit population at time t:

$$ R(t) = R_{0}e^{rt} $$

we will instead model this with a discrete iteration.  In Python, we can create an infinite iteration using a generator:


````python
def rabbits(R, birthrate, deathrate):
    """Creates a generator that produces an infinite supply of rabbits.
        R: the initial population of rabbits
        birthrate & deathrate: proportion of rabbits that 
        reproduce every time step"""
    while True:
        r = (birthrate - deathrate)
        R += R*r
        yield R
````

````python
def plot_pop(gen, n_weeks, label='rabbits'):
    """Plot a population generator on standard axes."""
    w = range(n_weeks)
    pop = [next(gen) for i in w]
    plt.xlabel('t')
    plt.ylabel('population')
    plt.plot(w, pop, label=label)

max_t = 500
plot_pop(rabbits(10, 0.02, 0.01), max_t, 'more births')
plot_pop(rabbits(500, 0.01, 0.02), max_t, 'more deaths')
plot_pop(rabbits(300, 0.01, 0.01), max_t, 'stable')
plt.legend(loc='upper left')

````

This seems a little bit unlikely, where are all these rabbits going to live?  Suppose that the habitat can only support a maximum number of rabbits (k).  The rate of growth in the rabbit population will become the *logistic equation*:

$$ \frac{dR}{dt} = \frac{rR(R-k)}{k} $$

And we can write a new rabbit generator to take this into account with a new ```max_rabbits``` parameter:

````python
def rabbits2(R, birthrate, deathrate, max_rabbits):
    """Creates a generator that produces an supply of rabbits.
        R: the initial population of rabbits
        birthrate & deathrate: proportion of rabbits that 
        reproduce or die every time step
        max_rabbits: the maximum sustainable population of rabbits."""
    R = float(R)
    while True:
        r = birthrate - deathrate
        k = (max_rabbits - R) / max_rabbits
        R += R*k*r
        yield R
        
````

````python
max_t = 100
plot_pop(rabbits2(10, 0.2, 0.1, 500), max_t)
plot_pop(rabbits2(750, 0.2, 0.1, 500), max_t, 'above max pop')
legend(loc='lower right')
````

## Foxes

Foxes eat rabbits. When there are rabbits to eat, foxes can reproduce. When there aren't enough rabbits then the foxes will die out.  This leads to the _Lotka-Volterra_ coupled equations:

### Rabbit Population

$$ \frac{dR}{dt} = aR - bFR $$

Where ```a``` is the basic birthrate of rabbits in the absence of foxes.  And ```b``` is the evasive ability of rabbits to avoid being eaten by foxes.

### Fox Population

$$ \frac{dF}{dt} = cRF - dF $$

Where ```c``` is the efficiency of the foxes in catching rabbits and converting them into new foxes.  And ```d``` is the deathrate of foxes.

````python
def foxes(F, rabbits_to_foxes, deathrate):
    """Creates a fox generator.
    As the population of foxes depends on the number of rabbits, we will need to
    send the generator the number of rabbits available at each iteration.
        F: Initial population of foxes
        rabbits_to_foxes: The conversion efficiency of rabbits to foxes
        deathrate:  the background deathrate of foxes"""
    F = float(F)
    while True:
        rabbits = yield F
        F += F*(rabbits_to_foxes*rabbits - deathrate)

        
def rabbits3(R, birthrate, evasion):
    """Creates a rabbit generator.
    As the population of rabbits depends on the number of foxes, we will need to
    send the generator the number of foxes predating them at each iteration.
        R: Initial population of rabbits
        birthrate: the birthrate of new rabbits
        evasion:  the chances of a rabbit being eaten by a fox"""
    R = float(R)
    while True:
        foxes = yield R
        R += R*(birthrate - foxes*evasion)
        
f = foxes(20, 0.0003, 0.01)
r = rabbits3(60, 0.015, 0.0005)
n_foxes = [next(f)]
n_rabbits = [next(r)]

iters = 2000
for i in range(iters):
    n_foxes.append(f.send(n_rabbits[-1]))
    n_rabbits.append(r.send(n_foxes[-2]))


plot(range(iters+1), n_foxes, label='foxes')
plot(range(iters+1), n_rabbits, label='rabbits')
legend(loc='upper left')
````

Small changes to the inputs can make big differences to the population cycles.  Try changing the variables in the cell below and rerunning with Shift-Enter.

````python
weeks     =  range(2000)
ecosystem = {'rabbits': [60],
             'foxes':   [20]}

params    = {'rabbit_birthrate': 0.015,
             'rabbit_evasion':   0.0005,  # The chances of a rabbit being eaten by _one_ fox
             'fox_deathrate':    0.01,
             'fox_efficiency':   0.0003}

def breed_rabbits(R, F, evasion, birthrate):
    return R + R*(birthrate - F*evasion)

def breed_foxes(F, R, efficiency, deathrate):
    return F + F*(R*efficiency - deathrate)

for w in weeks[:-1]:
    r, f = ecosystem['rabbits'][-1], ecosystem['foxes'][-1]
    ecosystem['rabbits'].append(breed_rabbits(r, f, params['rabbit_evasion'], params['rabbit_birthrate']))
    ecosystem['foxes'].append(breed_foxes(f, r, params['fox_efficiency'], params['fox_deathrate']))

for species, pop in ecosystem.items():
    plot(weeks, pop, label=species)

legend(loc='upper left')
````
